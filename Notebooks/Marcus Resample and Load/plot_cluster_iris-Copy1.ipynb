
# K-means Clustering


The plots display firstly what a K-means algorithm would yield
using three clusters. It is then shown what the effect of a bad
initialization is on the classification process:
By setting n_init to only 1 (default is 10), the amount of
times that the algorithm will be run with different centroid
seeds is reduced.
The next plot displays what using eight clusters would deliver
and finally the ground truth.




```python
print(__doc__)


# Code source: Gaël Varoquaux
# Modified for documentation by Jaques Grobler
# License: BSD 3 clause

import numpy as np
import matplotlib.pyplot as plt
# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import KMeans
from sklearn import datasets

np.random.seed(5)

iris = datasets.load_iris()
X = iris.data
y = iris.target


myyellow = [255/255,192/255,0]
myblue = [0/255,112/255, 192/255]

# Plot the ground truth
fig = plt.figure(1, figsize=(15, 10))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=30, azim=130)

for name, label in [('Single homes', 0),
                    ('Commerical', 1),
                    ('Industrial', 2)]:
    ax.text3D(X[y == label, 3].mean(),
              X[y == label, 0].mean(),
              X[y == label, 2].mean() + 2, name,
              horizontalalignment='center',
              bbox=dict(alpha=.2, edgecolor='w', facecolor='w'),
             size=20,
             )
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X[:, 3], X[:, 0], X[:, 2], s=200,c=y, edgecolor='k')

fsize = 20

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([],myyellow)
ax.set_xlabel('[PC1] "Frequency-like"',fontsize=fsize,color=myyellow)
ax.set_ylabel('[PC2] "Consumption-like"',fontsize=fsize,color=myyellow)
ax.set_zlabel('[PC3] "Behaviour-like"',fontsize=fsize,color=myyellow)
#ax.set_title('Ground Truth')
ax.dist = 12

ax.set_facecolor(myblue)

fig.show()
fig.savefig('clustering.png',bbox_inches='tight')
```
